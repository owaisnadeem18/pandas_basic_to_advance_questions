## 🔍 GroupBy: Split-Apply-Combine Paradigm

| Step | Description | Output |
|------|-------------|--------|
| Split | Group data by column(s) | `DataFrameGroupBy` object |
| Apply | Run aggregation/transformation/filter | Group results |
| Combine | Merge into final structure | Tidy DataFrame |

### 📐 Named Aggregation

```python
df.head()
```

```python
df.groupby("segment").agg(
    total_sales=("sales" , "sum"),
    avg_discount=("discount" , "mean"),
    order_count=("order_id" , "count")
)
```

### 🧪 Custom Aggregation

```python
df.groupby( "region")["profit"].agg(
    mean_profit = "mean",
    profit_range = lambda x: x.max() - x.min() 
    )
```

### 🔁 Transform Example (Z-score)

```python
df["sales_zscore"] = df.groupby("category")["sales"].transform(
    lambda x: (x - x.mean()) / x.std()
)
df[["category", "sales", "sales_zscore"]].head()
```

```python
region_summary = df.groupby("region").agg(
    total_sales = ("sales" , "sum"),
    avg_quantity = ("quantity" , "mean")
)

region_summary
```

```python
# do practice of question task # 01 again:

# 1. Find total sales and average quantity for each `region`.

region_summ = df.groupby("region").agg(
    total_sales = ("sales" , "sum" ),
    quantity = ("quantity" , "mean" )
)

region_summ
```

2. For each `category`, compute the standard deviation and CV of `sales`.

```python
category_summary = df.groupby("category").agg(
    sales_std = ("sales", "std"),
    sales_cv = ("sales" , lambda x: x.std() / x.mean()) # coefficeint formula is Std Deviation / mean
)

category_summary
```

3. Create a column `high_discount` = `True` if `discount` > average discount for its segment.

```python
df.head()
```

```python
avg_discount = df.groupby('segment')["discount"].transform("mean")

df["high_discount"] = (
    df["discount"] > avg_discount
)

avg_discount
```

```python
df
```

4. Filter only those sub-categories where total `sales` > 10000.

```python
# It gives the total summary of sales of each sub-category:
#  df.groupby("sub_category")["sales"].sum()

df.groupby("sub_category").filter(
    lambda x: x["sales"].sum() > 10000   
)
```

5. Normalize `profit` within each category using min-max scaling

```python
df["normalized_profit"] = df.groupby("category")["profit"].transform(
    lambda x: (x - x.min()) / (x.max() - x.min())
)
```

```python
import matplotlib.pyplot as plt

for region, data in df.groupby("region"):
    plt.plot(
        data["order_date"],
        data["sales"],
        label=region
    )

plt.xlabel("Date")
plt.ylabel("Cumulative Sales")
plt.title("Cumulative Sales per Region Over Time")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
