# **🧹 4. Data Cleaning**

```python
df_clean = df.copy()
df_clean.drop(["deck"] , axis=1 , inplace=True)
```

```python
df_clean
```

```python
df_clean.rename(columns={'sibsp' : "siblings_spouses_aboard"} , inplace=True )
```

## ❓ 5. 'Missing Values' Handling

```python
df_clean.isnull().sum()
```

```python
df_clean["age"].fillna(df["age"].median() , inplace= True)
```

```python
df_clean.isnull().sum()
```

```python
df_clean.head()
```

```python
df_clean.dropna(subset=["embark_town"] , inplace=True)
```

```python
df_clean.isnull().sum()
```

# **Fill 'embark_town' with its most frequent value**

```python
mode_value = df_clean['embark_town'].mode()[0]
df_clean['embark_town'] = df_clean['embark_town'].fillna(mode_value)
```

```python
df_clean["embark_town"].value_counts(normalize=True) * 100
```

# **🛠️ 6. Feature Engineering**

```python
df_clean["family_size"] = df_clean["siblings_spouses_aboard"] + df_clean["parch"] + 1 
df_clean["is_child"] = df_clean["age"].apply(lambda child : child < 12)
df_clean['sex_code'] = df_clean["sex"].map({"female" : 0 , "male" : 1 })
```

```python
df_clean.head()
```

# **Create column is_elderly for age > 65**

```python
df_clean["is_elderly"] = df_clean["age"].apply(lambda x: x > 60)
```

```python
df_clean.head()
```

# **Aggregation & Grouping**

```python
df_clean['sex'].value_counts()

df_clean.groupby('sex')['survived'].mean()

df_clean.groupby(["class" , "sex"])["age"].agg(["mean" , "count"])
```
